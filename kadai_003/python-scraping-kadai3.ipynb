```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from getpass import getpass
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_driver = webdriver.Chrome(options=chrome_options)
chrome_driver.get('https://terakoya.sejuku.net')

wait = WebDriverWait(chrome_driver, 30)
wait.until(EC.presence_of_all_elements_located)
header_login_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, 
         '#root > header > div > div > div.sc-bYMpWt.ecQgNh >'
         'div.sc-jIRcFI.sc-hhOBVt.gREswC.jPXpMh')
    )
)
header_login_button.click()

email_address = input('メールアドレスを入力してください: ')
password = getpass('パスワードを入力してください: ')
email_input = chrome_driver.find_element(By.NAME, 'email')
password_input = chrome_driver.find_element(By.NAME, 'password')
email_input.send_keys(email_address)
password_input.send_keys(password)
form_login_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR,
         '#root >.sc-pyfCe >.sc-kDvujY >.sc-eDWCr >button')
    )
)
form_login_button.click()
wait.until(EC.presence_of_all_elements_located)
import time
time.sleep(5)

chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/header/div[2]/div[2]/div/img').click()
time.sleep(2)
chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/header/div[3]/ul/a[1]/li').click()
time.sleep(2)
chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/main/div/div[1]/button').click()
time.sleep(2)
chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/main/div/div[1]/div[14]/div[2]/textarea').clear()
chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/main/div/div[1]/div[14]/div[2]/textarea').send_keys("プログラミング学習中です！今はスクレイピングに挑戦しています！")
time.sleep(2)
chrome_driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div/div/main/div/div[1]/button[2]').click()
time.sleep(2)
chrome_driver.quit()

```
